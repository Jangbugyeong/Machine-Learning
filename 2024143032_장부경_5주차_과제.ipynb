```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python

x = np.linspace(0, np.pi, 10)
y = np.linspace(0, np.pi, 10)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) + np.sin(Y)

plt.imshow(Z, extent=(0, 10, 0, 10), origin='lower', cmap='viridis')

plt.title('10x10 Heatmap of sin(x) + sin(y)')


plt.xticks(np.linspace(0, 10, 6))  # 0, 2, 4, 6, 8, 10
plt.yticks(np.linspace(0, 10, 6))

plt.show()
```

```python

x = np.linspace(0, np.pi, 100)
X = np.meshgrid(x)


Z = np.sin(X)


plt.imshow(Z, extent=(0, 100, 0, 100))
plt.colorbar(label='sin(x) + sin(y)')
plt.title('100x100 Heatmap ')



plt.xticks(np.linspace(0, 100, 6))
plt.yticks(np.linspace(0, 100, 6))

plt.show()
```
